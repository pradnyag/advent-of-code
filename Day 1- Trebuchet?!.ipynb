```python
import pandas as pd 
df = pd.read_csv("input",  header=None, names=["text"])

# Rename the column to "text"
df.columns = ["text"]
```

# Part 1

```python
string = "treb7uchet"
output = ""

for i in range(len(string)):
    if(string[i].isdigit()):
        output += str(string[i])
        break
    
for i in range(len(string) -1, -1, -1):
    if(string[i].isdigit()):
        output += str(string[i])
        break

print(int(output))
```

```python
df['calibration_value'] = 0 

for index, row in df.iterrows():

    output = ""
    string = row["text"]

    for i in range(len(string)):
        if(string[i].isdigit()):
            output += str(string[i])
            break
    
    for i in range(len(string) -1, -1, -1):
        if(string[i].isdigit()):
            output += str(string[i])
            break

    df.at[index, 'calibration_value'] = int(output)
```
