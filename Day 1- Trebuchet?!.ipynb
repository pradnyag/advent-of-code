```python
df['calibration_value'] = 0 

for index, row in df.iterrows():

    output = ""
    string = row["text"]

    for i in range(len(string)):
        if(string[i].isdigit()):
            output += str(string[i])
            break
    
    for i in range(len(string) -1, -1, -1):
        if(string[i].isdigit()):
            output += str(string[i])
            break

    df.at[index, 'calibration_value'] = int(output)
```

# Part 2

```python
df['updated_calibration_value'] = 0 

hashmap = {"zero" : "0" , "one" : "1", "two" : "2", "three" : "3", "four": "4","five" : "5", "six" : "6", "seven": "7" , "eight": "8", "nine" : "9" }

for index, row in df.iterrows():

    string = row["text"]
    output = ""
    substring = ""
    counter = 0

    # Forward pass
    for i in range(len(string)):

        substring += string[i]

        if ((string[i].isdigit())):
            output += string[i]
            break

        for key in hashmap.keys():
            if (key in substring):
                output += hashmap[key]
                substring = ""
                counter = 1
                break
        
        if (counter == 1):
            break

    # Reinitialising the counter
    counter = 0
     
    # Backward pass
    for i in range(len(string) - 1, -1, -1):

        substring += string[i]
        
        if ((string[i].isdigit())):
            output += string[i]
            break

        for key in hashmap.keys():
            if (key in substring):
                output += hashmap[key]
                substring = ""
                counter = 1
                break
        
        if (counter == 1):
            break

        
    df.at[index,'updated_calibration_value']= int(output)
    
```
